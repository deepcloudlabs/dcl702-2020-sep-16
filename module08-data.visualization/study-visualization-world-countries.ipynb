```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
countries = pd.read_json("countries.json")
countries2 = countries.dropna()
countries3 = countries2.drop(columns=["cities","localName","capital"])
countries3.rename(columns={"_id": "code"}, inplace=True)
```

```python
countries3
```

```python
fig = plt.figure()
countries3["lifeExpectancy"].plot()
plt.legend(loc="best")
```

```python
fig = plt.figure()
plt.bar(x = "name" , data=countries3, height = 'lifeExpectancy')
```

```python
fig = plt.figure()
sns.histplot(data=countries3, x= 'continent')
```

```python
fig = plt.figure()
sns.pairplot(data=countries3)
```

```python
fig = plt.figure()
sns.regplot(x='lifeExpectancy', y='gnp', data=countries3)
```
