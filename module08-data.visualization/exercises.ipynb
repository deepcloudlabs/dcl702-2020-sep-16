```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(np.arange(10),'ro--',label='USD')
ax1.plot(np.arange(10)*2,'g',label='EUR')
ax1.legend(loc='best')
ax2.plot(np.random.randn(50).cumsum(), color = '#A0A0A0' , linestyle='--')
ax3.hist(np.random.randn(100),bins=20, color='r', alpha=0.5 )
ax4.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
fig, axes = plt.subplots(2,3)
```

```python
for i in range(2):
    for j in range(3):
        axes[i,j].hist(np.random.randn(500),bins=50, color='g', alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=0)        
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), 'r' , label = "ORCL")
ax.plot(np.random.randn(1000).cumsum(), 'g', label = "MSFT")
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticklabels(["one", "two", "three", "four", "five"], rotation=90, fontsize='small')
ax.set_xlabel('Levels')
ax.set_title("ORCL Stock price changes last week")
ax.legend(loc="best")
```
