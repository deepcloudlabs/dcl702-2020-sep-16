```python
import pandas as pd
import numpy as np
```

```python
frame1 = pd.DataFrame({
    "key": ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
    "data1": range(7) 
})
```

```python
frame2 = pd.DataFrame({
    "key": ['a', 'b', 'd'],
    "data2": range(3) 
})
```

```python
frame1
```

```python
frame2
```

```python
pd.merge(frame1,frame2,on='key')
```

```python
frame3 = pd.DataFrame({
    "key1": ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
    "data1": range(7) 
})
```

```python
frame4 = pd.DataFrame({
    "key2": ['a', 'b', 'd'],
    "data2": range(3) 
})
```

```python
frame3
```

```python
frame4
```

```python
pd.merge(frame3,frame4,left_on="key1",right_on="key2")
```

```python
pd.merge(frame1,frame2, on="key",how="left")
```

```python
pd.merge(frame1,frame2, on="key",how="right")
```

```python
pd.merge(frame1,frame2, on="key",how="inner")
```

```python
pd.merge(frame1,frame2, on="key", how="outer")
```

```python
frame5 = pd.DataFrame({
    "key1": ['jack', 'jack', 'kate'],
    "key2": ['shephard', 'bauer', 'shephard'],
    "data1": [37, 56, 32]
})
```

```python
frame6 = pd.DataFrame({
    "key1": ['jack', 'jack', 'kate', 'kate'],
    "key2": ['shephard', 'shephard', 'shephard', 'bauer'],
    "data2": [43, 58, 36, 41 ]
})
```

```python
frame5
```

```python
frame6
```

```python
pd.merge(frame5,frame6, on=["key1", "key2"], how="outer")
```

```python
pd.merge(frame5,frame6, on="key1")
```

```python
pd.merge(frame5,frame6, on="key1", suffixes=('_left', '_right'))
```
